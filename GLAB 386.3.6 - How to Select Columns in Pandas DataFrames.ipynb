##**Introduction:**


**Viewing Column Names Using Column Attribute**

- Get all column names:
  ```
  df.columns
  ```

- Get a single column name by index position:
  ```
  df.columns[1]
    ```

- Get multiple column names by index positions:
  ```
  df.columns[[0, 3, 4]]
    ```

**Selecting Columns Using Square Brackets `[ ]`**

- To select a **single column**, use square brackets [ ] with the column name:
  ```   
  df['colName']
    ```

- To select **multiple columns**, use a list of column names in the down square bracker:  
  ```
  df[['col1', 'col2']]

  ```

  **Simple Way to Remember**

- df['colName'] → gives you the data

- df.columns[0] → gives you the name

## **Dataset:**

The lab utilizes the [`cars.json`](https://drive.google.com/file/d/1CXAK8gbuLtc2NNOXVUgmja8fDg0TrNZm/view) dataset, providing a practical context for applying column selection methods.

 Let import cars dataset in Panda dataframe.

````python
import pandas as pd
df_cars = pd.read_json('files/cars.json')
df_cars.head()
````

##**Example: View Single Column Name by Index Number**

````python
df_cars.columns[0] # first column of the DataFrame
````

##**Example: View Multiple Column Name by Index Number**

````python
df_cars.columns[[0,1,9]] # first,sencond,8th columns of the DataFrame by index  
````

##**Example - Select Single Column Data using Square brackets [ ]:**

Suppose we are interested in the Name of the cars.

````python
df_cars['Car'] # Column named 'Cars'
````

````python
df_cars[df_cars.columns[0]] #first column of the DataFrame
````

    If you notice, both code examples give the same result because:

- `df_cars['Car']`:  
  This directly selects the column named **'Car'** and returns the column data as a **Series**.  
  This is called **direct column access**.

- `df_cars[df_cars.columns[0]]`:  
  This first gets the name of the first column using `df_cars.columns[0]`, which is `'Car'`.  
  It then uses that name to select the same column data.  
  This is called **dynamic column selection by position**.

Since:
`df_cars.columns[0] == 'Car'`

Both statements return the same result.

- `df_cars['Car']` → hardcoded column name  
- `df_cars[df_cars.columns[0]]` → dynamic column selection by position

##**Example: Select Multiple Columns Data by Indexes**

````python
df_cars[df_cars.columns[[0,1,9]]] #first,second and 8th columns of the DataFrame
````



---



##**Example: Select multiple Columns Data using Square brackets [ ]**

Suppose, we are interested in the cars name, Model of the cars, and quantity

````python
df_cars[['Car','Model','quantity']] # Columns named 'Car', 'Model', and 'quantity'
````
