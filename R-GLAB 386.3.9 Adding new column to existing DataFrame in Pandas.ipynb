## **Method #2: By using DataFrame.insert() function**

It gives the freedom to add a column at any position we like, and not just at the end. It also provides different options for inserting the column values.
- **Modifies In-Place:** insert() modifies the DataFrame in place. It does not return a new DataFrame; it directly changes the existing one.

- **One Column at a Time:** insert() is designed to add a single column at a time. You cannot add multiple columns in a single insert() call.
- **No Direct Chaining: **Due to its in-place modification, insert() is not typically used for method chaining in the same way assign() is.
- **Handles Duplicates:** It raises a ValueError if the column name already exists, unless allow_duplicates=True is set (which is generally not recommended for clarity).

## **When to use insert():**

- When you need to insert a column at a specific position (e.g., at the beginning or in the middle of existing columns).
- When you want to modify the original DataFrame directly and don't need a new copy.
- When you are adding a single column.


**Example 2.1**

In this example, we will convert a Python dictionary into a Pandas DataFrame and utilize the insert() method to add a new "Age" column at a specific index. This highlights how to structure labeled datasets and precisely control the placement of new data within an existing table.


```python
# Import pandas package
import pandas as pd

# Define a dictionary containing Students data
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
       'Score 1' : [56,86,77,45],
           'Score 2' : [50,96,60,30]}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
print("------before -------")
print(df)
# Using DataFrame.insert() adding age column
df.insert(2, "Age", [21, 23, 24, 21], True)
print("------after adding column -------")
# Observe the result
print(df)
```

## **Method #3: Using Dataframe.assign() function**

**Example 3.1**


- **Returns a New DataFrame (Non-Inplace):** The assign() method returns a new DataFrame with the new column(s) added. It does not modify the original DataFrame in place. If you want to keep the changes, you must assign the result back to a variable (e.g., df = df.assign(...)).
- **Method Chaining:** Because it returns a new DataFrame, assign() is excellent for method chaining. You can create multiple new columns or perform other DataFrame operations in a single, readable line.
- **Creating Multiple Columns:** You can create multiple new columns simultaneously by passing multiple keyword arguments to assign().
Referencing Other New Columns: In Python 3.6 and later, you can reference newly created or modified columns within the same assign() call. This allows for dependencies between the new columns.
- **Using Callables (Lambda Functions):** assign() often uses lambda functions (or other callables) to define the new column's values based on existing columns in the DataFrame. This makes it very flexible for calculated columns.

## **When to use assign():**
- When you want to create a new DataFrame without modifying the original.
- When you are doing a series of transformations and want to use method chaining for readability.
- When you need to create multiple new columns at once.
- When the new column's values are derived from existing columns through calculations or logic.

```python
# Import pandas package
import pandas as pd

# Define a dictionary containing Students data
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
       'Score 1' : [56,86,77,45],
           'Score 2' : [50,96,60,30]}

print("------before -------")

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
print(df)
print("------after adding column -------")
# using DataFrame.assign() method adding 'Address' as the column name and equating it to the list

df = df.assign(address = ['NYC', 'NJ', 'CA', 'PA'])

# Observe the result
print(df)
```
